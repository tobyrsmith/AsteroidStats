```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
AS = np.loadtxt('AllAst.dat', unpack=True)

ASP = AS ** 1.5
```

```python
TNO = np.loadtxt('TNO_a.csv', unpack=True)

TNOP = TNO ** 1.5
```

```python
print TNO.size
```

```python
def log_10_product(x, pos):
    if (x < 1.0):
      return '%4.2f' % (x)
    else:
      return '%i' % (x)
```

```python
tfont = {
    'family' : 'DejaVu Serif',
    'weight' : 'bold',
    'size' : 14}

plt.rc('font', **tfont)
plt.rc("axes", linewidth=3.0)
plt.rc('axes', grid=False)

formatter = plt.FuncFormatter(log_10_product)
```

```python
# Make bins

Bins = np.linspace(0,14,1000)
Width = (Bins[1:] - Bins[:-1])
```

```python
print Bins.size
print Width.size
```

```python
ASHist,BinEdge = np.histogram(ASP, bins=Bins)
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(10,7)

#ax.grid(b=True, which='major')

#ax.set_ylim(0,7000)
ax.set_xlim(1.5,13)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

ax.set_xlabel('Orbital Period [years]')
ax.set_ylabel('Number of Asteroids')

ax.bar(Bins[:-1], ASHist, width=Width, color='r', edgecolor='r');
```

```python
fig.savefig('AstHist_Period.png', dpi=300, bbox_inches='tight',pad_inches=0.05)
```

```python
ASJ = 11.85920 / (AS ** 1.5)
```

```python
Bins = np.linspace(0,5,500)
Width = (Bins[1:] - Bins[:-1])

ASJHist,BinEdge = np.histogram(ASJ, bins=Bins)
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(10,7)

#ax.grid(b=True, which='major')

#ax.set_ylim(0,7000)
ax.set_xlim(4.75,0.75)

ax.set_xticks([1, 2, 3, 4])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

ax.set_xlabel('Number of Orbits for Every 1 Jupiter Orbit')
ax.set_ylabel('Number of Asteroids')

ax.bar(Bins[:-1], ASJHist, width=Width, color='r', edgecolor='r');
```

```python
fig.savefig('AstHist_PeriodJ.png', dpi=300, bbox_inches='tight',pad_inches=0.05)
```
